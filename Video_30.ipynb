```python
import pandas as pd
import numpy as np
# Datos de ejemplo 
ventas = pd.DataFrame({ 'fecha': pd.date_range('2024-01-01', periods=100, freq='D'), 'producto': np.random.choice(['Laptop', 'Mouse', 'Teclado'], 100), 'cantidad': np.random.randint(1, 10, 100), 'precio': np.random.uniform(10, 1000, 100) }) 
```

```python
#conociendo datos
print(ventas.info())
```

```python
print(ventas.describe())
```

```python
print(ventas.isnull().sum())
```

```python
#nuevas variables
ventas['total']=ventas['cantidad']*ventas['precio']
ventas['mes']=ventas['fecha'].dt.month
```

```python
#Preguntas de negocio
print('1. Cual es el producto mas vendido?')
print(ventas.groupby('producto')['cantidad'].sum().sort_values(ascending=False))
```

```python
#2. ingresos por mes
print('\n2.Cuales son los ingresos por mes?')
print(ventas.groupby('mes')['total'].sum())
```

```python
#3ticket promedio
print('\n3.Cual es el ticket promedio por producto?')
print(ventas.groupby('producto')['total'].mean().round(2))
```
