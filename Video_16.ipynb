```python
#Tabla de multiplicar

print("TABLA DE MULTIPLICAR: GENERADOR")

numero=int(input("De que numero quieres la tabla?"))
print(f"\n Tabla del {numero}:")

for i in range(1,11):
    resultado=numero *i
    print(f"{numero}x{i:2d}={resultado:3d}")

# Crear mÃºltiples tablas

print(f"\nðŸ“š TABLAS DEL 1 AL 5:")
for tabla in range(1, 6):
    print(f"\n--- Tabla del {tabla} ---")
    for multiplicador in range(1, 6):
        print(f"{tabla} x {multiplicador} = {tabla * multiplicador:2d}")
```

```python
#Analisis de Datos
print("ANALISIS DE VENTAS SEMANALES")

#simular recoleccion de datos de ventas
ventas_semana=[]
dias_semana=["Lunes","Martes","Miercoles","Jueves","Viernes","Sabado","Domingo"]

print("Ingresa las ventas de  cada dia:")
for i in range(7):
    venta=float(input(f"Ventas del {dias_semana[i]}:$")
    ventas_semana.append(venta)

# AnÃ¡lisis de datos
total_ventas = 0
mejor_dia = ""
mejor_venta = 0
peor_dia = ""
peor_venta = float('inf')  # Infinito para comparaciÃ³n

print(f"\nðŸ“Š === RESUMEN SEMANAL ===")
for i in range(7):
    venta_dia = ventas_semana[i]
    total_ventas += venta_dia
    
    print(f"{dias_semana[i]:<10}: ${venta_dia:8.2f}")
    
    # Encontrar mejor dÃ­a
    if venta_dia > mejor_venta:
        mejor_venta = venta_dia
        mejor_dia = dias_semana[i]
    
    # Encontrar peor dÃ­a
    if venta_dia < peor_venta:
        peor_venta = venta_dia
        peor_dia = dias_semana[i]

promedio_diario=total_ventas/7

print(f"\n Total ventas:${total_ventas:.2f}")
print(f"Promedio diario:${promedio_diario:.2f}")
print(f"Mejor dia:{mejor_dia}(${mejor_venta:.2f})")
print(f"Peor dia:{peor_dia}(${peor_venta:.2f})")
```
