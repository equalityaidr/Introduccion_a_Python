```python
def analizar_calificaciones(calificaciones):
    if not calificaciones:
        return 0,0,0,0, "Sin datos"

    promedio=sum(calificaciones)/len(calificaciones)
    maximo=max(calificaciones)
    minimo=min(calificaciones)
    total=len(calificaciones)
   
    # Determinar estado académico
    if promedio >= 90:
        estado = "Excelente"
    elif promedio >= 80:
        estado = "Muy bueno"
    elif promedio >= 70:
        estado = "Bueno"
    elif promedio >= 60:
        estado = "Regular"
    else:
        estado = "Necesita apoyo"
    
    return promedio, maximo, minimo, total, estado

#Usar la funcion
estudiantes={"Ana":[85,90,88,92,87],"Carlos":[95,98,92,94,96]}

print("ANALISIS DE CALIFICACIONES")

for nombre, notas in estudiantes.items():
    prom, max_nota,min_nota,total_materias,estado=analizar_calificaciones(notas)

print(f"\n{nombre}:")
print(f"{total_materias}")
print(f"{prom:.2f}")
print(f"{max_nota}")
print(f"{min_nota}")
print(f"{estado}")

```

```python
def celsius_a_fahrenheit(celsius):
    return (celsius*9/5)+32

def fahrenheit_a_celsius(fahrenheit):
    return(fahrenheit-32)*5/9

def clasificar_temperatura(celsius):
    if celsius<0:
        return "Muy frio (congelado)"
    elif celsius<10:
        return "Frio"

    elif celsius<20:
        return "Fresco"
    elif celsius<30:
        return "Agradable"
    elif celsius<40:
        return "Calor"
    else:
        return "Muy caliente"

def reporte_temperatura_completo(temperatura, unidad):
    """
    Genera reporte completo de temperatura
    
    Args:
        temperatura (float): Valor de temperatura
        unidad (str): 'C' para Celsius, 'F' para Fahrenheit
    
    Returns:
        dict: Reporte completo con conversiones y clasificación
    """
    if unidad.upper() == 'C':
        celsius = temperatura
        fahrenheit = celsius_a_fahrenheit(temperatura)
    elif unidad.upper() == 'F':
        fahrenheit = temperatura
        celsius = fahrenheit_a_celsius(temperatura)
    else:
        return {"error": "Unidad debe ser 'C' o 'F'"}
    
    clasificacion = clasificar_temperatura(celsius)
    
    return {
        "celsius": round(celsius, 2),
        "fahrenheit": round(fahrenheit, 2),
        "clasificacion": clasificacion,
        "unidad_original": unidad.upper()
    }

# Probar el sistema de temperaturas
temperaturas_test=[(25,'C'),(77,'F'),(0,'C')]

print("REPORTE DE TEMPERATURAS")

for temp, unidad in temperaturas_test:
    reporte=reporte_temperatura_completo(temp,unidad)
    if "error" not in reporte:
        print(f"Temperatura original:{temp}'{reporte['unidad_original']}")
        print(f"Celsius:{reporte['celsius']}C")
        print(f"Fahrenheit:{reporte['fahrenheit']}F")
        print(f"Clasificacion:{reporte['clasificacion']}")
        print()
        
```
