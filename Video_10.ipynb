```python
# De número a texto 
edad = 25 
edad_texto = str(edad) 
print(f"Mi edad es {edad_texto}") # Funciona! 
print(type(edad_texto)) # <class 'str'> 
```

```python
# De booleano a texto 
activo = True 
estado_texto = str(activo) 
print(f"Estado: {estado_texto}") # "True"
```

```python
# De texto a entero 
precio_texto = "100" 
precio_numero = int(precio_texto) 
nuevo_precio = precio_numero + 50 
print(f"Precio con recargo: {nuevo_precio}") 

# De decimal a entero (¡cuidado!) 
precio_decimal = 19.99 
precio_entero = int(precio_decimal) # Se convierte en 19 (pierde decimales) print(precio_entero)
```

```python
# De texto a decimal 
altura_texto = "1.75" 
altura_numero = float(altura_texto)
 print(f"Altura: {altura_numero} metros") 
# De entero a decimal 
edad = 25 
edad_decimal = float(edad) # 25.0 
print(edad_decimal)
```
