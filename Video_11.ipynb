```python
edad=25
print(edad > 18) # True - es mayor que 18 
print(edad < 30) # True - es menor que 30 
print(edad == 25) # True - es igual a 25
print(edad != 30) # True - es diferente de 30 
print(edad >= 25) # True - es mayor o igual a 25 
print(edad <= 20) # False - NO es menor o igual a 20 
```

```python
print("===SISTEMA DE ACCESO===")
#Datos de usuario
usuario_ingresado=input("Usuario: ")
password_ingresada=input("Contraseña: ")
intentos_fallidos=int(input("Cuantos intentos fallidos has tenido?"))

#Datos correctos del sistema
usuario_correcto="admin"
password_correcto="python123"
max_intentos=3

#Logica de autenticacion
credenciales_correctas=(usuario_ingresado==usuario_correcto and password_ingresada==password_correcto)
no_bloqueado=intentos_fallidos<max_intentos
acceso_permitido=credenciales_correctas and no_bloqueado

#resultados
print(f"\n===RESULTADO===")
print(f"Credenciales correctas:{credenciales_correctas}")
print(f"No bloqueado:{no_bloqueado}")
print(f"Acceso permitido:{acceso_permitido}") 
if acceso_permitido:
    print("Bienvenido al sistema!")
else: print("Acceso denegado")

# Repaso rápido de operadores lógicos
print("=== TABLA DE VERDAD ===") 
print(f"True and True = {True and True}") # True
print(f"True and False = {True and False}") # False 
print(f"True or False = {True or False}") # True 
print(f"False or False = {False or False}") # False 
print(f"not True = {not True}") # False 
print(f"not False = {not False}") # True
```
