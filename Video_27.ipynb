```python
def dividir_seguro(a, b):
    try:
        resultado = a / b
        return resultado
    except ZeroDivisionError:
        print("Error: No se puede dividir por cero")
        return None

# Probar la funci√≥n
print("=== DIVISI√ìN SEGURA ===")

print(f"10/2={dividir_seguro(10,2)}")
print(f"15/3={dividir_seguro(15,3)}")
print(f"8/0={dividir_seguro(8,0)}")
```

```python
def calculadora_robusta():
    print("CALCULADORA ROBUSTA")
    while True:
    try:
            # Entrada de datos con posibles errores
        num1 = float(input("Primer n√∫mero (o 'salir' para terminar): "))
        num2 = float(input("Segundo n√∫mero: "))
        operacion = input("Operaci√≥n (+, -, *, /): ")
            
            # Realizar c√°lculo
        if operacion == "+":
            resultado = num1 + num2
        elif operacion == "-":
            resultado = num1 - num2
        elif operacion == "*":
            resultado = num1 * num2
        elif operacion == "/":
            resultado = num1 / num2  # Puede causar ZeroDivisionError
        else:
            print("‚ö†Ô∏è Operaci√≥n no v√°lida")
            continue
            
        print(f"‚úÖ Resultado: {num1} {operacion} {num2} = {resultado}")
    except ValueError:
        entrada=input("Quieres salir?(s/n): ")
        if entrada.lower()=='s':
            break
        else:
            print("Por favor ingresa numeros validos")
            continue

    except ZeroDivisionError:
        print("Error:Division por cero no permitida")

    except KeyboardInterrupt:
        print("\nPrograma interrumpido por el usuario")
        break

    except Exception as e:
        print(f"Error inesperado:{e}")

#calculadora_robusta()


```

```python
def leer_calificaciones_archivo(nombre_archivo):
    calificaciones=[]
    archivo=None
    try:
        print(f"üìÇ Intentando abrir archivo: {nombre_archivo}")
        archivo = open(nombre_archivo, 'r', encoding='utf-8')
        
        contenido = archivo.read()
        print(f"‚úÖ Archivo le√≠do exitosamente")
        
        # Procesar contenido
        lineas = contenido.strip().split('\n')
        for i, linea in enumerate(lineas, 1):
            try:
                calificacion = float(linea.strip())
                if 0 <= calificacion <= 100:
                    calificaciones.append(calificacion)
                else:
                    print(f"‚ö†Ô∏è L√≠nea {i}: Calificaci√≥n fuera de rango (0-100): {calificacion}")
            except ValueError:
                print(f"‚ö†Ô∏è L√≠nea {i}: No es un n√∫mero v√°lido: '{linea.strip()}'")
    
    except FileNotFoundError:
        print(f"‚ùå Error: El archivo '{nombre_archivo}' no existe")
        # Crear archivo de ejemplo
        print("üí° Creando archivo de ejemplo...")
        with open('calificaciones_ejemplo.txt', 'w') as ejemplo:
            ejemplo.write("85\n90\n78\n92\n88\n")
        print("‚úÖ Archivo 'calificaciones_ejemplo.txt' creado")
        
    except PermissionError:
        print(f"‚ùå Error: Sin permisos para leer '{nombre_archivo}'")
        
    except UnicodeDecodeError:
        print(f"‚ùå Error: Problema de codificaci√≥n en '{nombre_archivo}'")
        
    else:
        # Solo se ejecuta si NO hubo errores en el try
        print(f"üéâ Procesamiento exitoso: {len(calificaciones)} calificaciones v√°lidas")
        
    finally:
        # SIEMPRE se ejecuta, haya o no errores
        if archivo and not archivo.closed:
            archivo.close()
            print("üìÅ Archivo cerrado correctamente")
        print("üîö Procesamiento completado")
    
    return calificaciones

#Demostracion
print("PROCESADOR DE CALIFICACIONES")

#probar archivo que no existe
califs1=leer_calificaciones_archivo("archivo_inexistente.txt")
print(f"Calificaciones obtenidas:{califs1}")
print()

#probar con archivo existente
califs2=leer_calificaciones_archivo("calificaciones_ejemplo.txt")
print(f"Calificaciones obtenidas:{califs2}")

if califs2:
    promedio=sum(califs2)/len(califs2)
    print(f"Promedio:{promedio:.2f}")
             
```
