```python
#Ejemplo pr√°ctico - Validaci√≥n de entrada:

print("VALIDADOR DE CONTRASE√ëA SEGURA")

password=""
intentos=0
max_intentos=3

while password != "python123" and intentos < max_intentos:
    password = input(f"Ingresa la contrase√±a (intento {intentos + 1}/{max_intentos}): ")
    intentos += 1
    
    if password != "python123":
        if intentos < max_intentos:
            print("‚ùå Contrase√±a incorrecta. Intenta de nuevo.")
        else:
            print("‚ùå Demasiados intentos fallidos. Acceso bloqueado.")
if password =="python123":
    print("Acceso concedido!")
```

```python
#Sistema de menu que sigue funcionando hasta que el usuario salga
print("SISTEMA DE PEDIDOS DE PIZZA")

total_pedido=0
seguir_ordenando=True

while seguir_ordenando:
    print(f"\nüí∞ Total actual: ${total_pedido:.2f}")
    print("üçï MEN√ö:")
    print("1. Pizza Margarita - $15.99")
    print("2. Pizza Pepperoni - $18.99") 
    print("3. Pizza Hawaiana - $19.99")
    print("4. Ver total y pagar")
    print("5. Cancelar pedido")
    
    opcion = input("\n¬øQu√© deseas hacer? (1-5): ")
    
    if opcion == "1":
        total_pedido += 15.99
        print("‚úÖ Pizza Margarita agregada al pedido")
    elif opcion == "2":
        total_pedido += 18.99
        print("‚úÖ Pizza Pepperoni agregada al pedido")
    elif opcion == "3":
        total_pedido += 19.99
        print("‚úÖ Pizza Hawaiana agregada al pedido")
    elif opcion == "4":
        if total_pedido > 0:
            print(f"\nüßæ TOTAL A PAGAR: ${total_pedido:.2f}")
            print("¬°Gracias por tu pedido!")
            seguir_ordenando = False
        else:
            print("‚ö†Ô∏è Tu carrito est√° vac√≠o")
    elif opcion=="5":
        print("Pedido Cancelado")
        seguir_ordenando=False
    else:
        print("Opcion no valida. Intenta denuevo")

print("Hasta la proxima!")

```
