```python
import pandas as pd
import matplotlib.pyplot as plt

#Cargar datos
datos=pd.read_csv('ventas_2024.csv')

#Analisis rapido
print(f'Total de ventas: ${datos["monto"].sum():,.2f}')
print(f'Promedio de venta: ${datos["monto"].mean():.2f}')

#Visualizacion instantanea
datos['monto'].hist(bins=20)
plt.title('Distribucion de Ventas')
plt.show()
```
