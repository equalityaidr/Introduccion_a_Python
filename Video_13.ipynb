```python
# Estructura completa con todos los elementos
edad = 20

if edad < 13:                    # 1. 'if' + condiciÃ³n + ':'
    print("Eres un niÃ±o")        # 2. IndentaciÃ³n de 4 espacios
    categoria = "niÃ±o"           # 3. Todo el bloque indentado igual
elif edad < 20:                  # 4. 'elif' + nueva condiciÃ³n + ':'
    print("Eres adolescente")    # 5. Mismo nivel de indentaciÃ³n
    categoria = "adolescente"
else:                           # 6. 'else' + ':' (sin condiciÃ³n)
    print("Eres adulto")        # 7. Mismo nivel de indentaciÃ³n
    categoria = "adulto"

print(f"CategorÃ­a final: {categoria}")  # 8. Sin indentaciÃ³n = fuera del condicional
```
