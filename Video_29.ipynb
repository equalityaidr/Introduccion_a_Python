```python
#Lista tradicional de Python
lista_python=[1,2,3,4,5]

#para multiplicar cada elemento por 2
lista_multiplicada=[x*2 for x in lista_python]

#Lista NumPy
import numpy as np
array_numpy=np.array([1,2,3,4,5])

#para multiplicar por 2
array_multiplicado=array_numpy*2
print(f'Lista python:{lista_multiplicada}')
print(f'Lista numpy:{array_multiplicado}')
```

```python
import numpy as np

# Datos de temperatura de 7 días 
temperaturas = np.array([ [22, 25, 28], # Día 1: mañana, tarde, noche 
                         [24, 27, 26], # Día 2 
                         [23, 29, 25], # Día 3 
                         [25, 31, 27], # Día 4 
                         [26, 30, 28], # Día 5 
                         [24, 28, 26], # Día 6 
                         [23, 26, 24] # Día 7 
                         ]) 

print(f'Temperatura promedio semanal:{temperaturas.mean():.1f}C')
print(f'Temperatura maxima:{temperaturas.max()}C')
print(f'Promedio por momento del dia:')
print(f'Mañanas:{temperaturas[:,0].mean():.1f}C')
print(f'Tardes:{temperaturas[:,1].mean():.1f}C')
print(f'Noches:{temperaturas[:,2].mean():.1f}C')
```
