```python
#Base de datos de productos
inventario={
    "P001":{"nombre":"laptop","precio":899.99,"stock":15},
    "P002":{"nombre":"mouse","precio":25.5,"stock":50},
    "P003":{"nombre":"teclado","precio":75.0,"stock":30},
    "P004":{"nombre":"monitor","precio":250.0,"stock":5}
}

print("SISTEMA DE INVENTARIO")

#keys- obtener todas las claves
print("Codigos de productos:",list(inventario.keys())

#value-obtener todos los valores
print("Informacion de productos:",list(inventario.values())

#items-obtener pares clave/valor

print("CATALOGO COMPLETO")

for codigo, info in iventario.items():
    print(f"Codigo:{codigo}")
    print(f"Producto:{info["nombre"]}")
    print(f"Precio:${info["precio"]}")
    print(f"Stock:${info["stock"]} unidades")  
```

```python
#base de datos de empresa
empresa={
    "empleados":
    {"E001":
     {"nombre":"Ana Garcia",
      "puesto":"Desarrolladora",
      "salario":45000,
      "departamento":"IT",
      "fecha_ingreso":"2022-03-15"
     },
     "E002":
     {"nombre":"Bruno Lopez",
      "puesto":"Dise√±ador",
      "salario":40000,
      "departamento":"Marketing",
      "fecha_ingreso":"2021-08-20"
     } ,
     "E003":
     {"nombre":"Carlos Ruiz",
      "puesto":"Manager",
      "salario":55000,
      "departamento":"IT",
      "fecha_ingreso":"2020-01-10"
     }},
    "departamentos":{"IT":{"presupuesto":150000,"ubicacion":"Piso 3"},
                     "Marketing":{"presupuesto":100000,"ubicacion":"Piso 2"}}}

print("SISTEMA DE RECURSOS HUMANOS")

#Acceso a informacion anidada

print("Informacion de Ana Garcia:")
ana=empresa["empleados"]["E001"]
print(f"Nombre:{ana["nombre"]}")
print(f"Puesto:{ana["puesto"]}")
print(f"Salario:${ana["salario"]}")

# An√°lisis por departamento
print(f"\nüìä AN√ÅLISIS POR DEPARTAMENTO:")
departamentos_empleados = {}

for emp_id, empleado in empresa["empleados"].items():
    dept = empleado["departamento"]
    if dept not in departamentos_empleados:
        departamentos_empleados[dept] = []
    departamentos_empleados[dept].append(empleado)

for dept, empleados_dept in departamentos_empleados.items():
    print(f"\n{dept}:")
    print(f"  Empleados: {len(empleados_dept)}")
    
    salarios = [emp["salario"] for emp in empleados_dept]
    print(f"  Salario promedio: ${sum(salarios) / len(salarios):,.2f}")
    print(f"  Presupuesto departamento: ${empresa['departamentos'][dept]['presupuesto']:,}")
    
    print("  Lista de empleados:")
    for emp in empleados_dept:
        print(f"    - {emp['nombre']} ({emp['puesto']})")
```
