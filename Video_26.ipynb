```python
#Usar una estructura de datos y pasar parametros

def crear_estado_juego_inicial(nombre_usuario):
    """Crea el estado inicial del juego"""
    return {
        "usuario": nombre_usuario,
        "puntuacion": 0,
        "nivel": 1,
        "vidas": 3,
        "items": [],
        "tiempo_inicio": "2024-01-15 10:30:00"
    }

def actualizar_puntuacion(estado_juego, puntos_ganados):
    """Actualiza la puntuaci贸n y devuelve nuevo estado"""
    nuevo_estado = estado_juego.copy()
    nuevo_estado["puntuacion"] += puntos_ganados
    
    # Verificar si sube de nivel
    if nuevo_estado["puntuacion"] >= nuevo_estado["nivel"] * 1000:
        nuevo_estado["nivel"] += 1
        print(f" 隆Subiste al nivel {nuevo_estado['nivel']}!")
    
    return nuevo_estado

def perder_vida(estado_juego):
    """Reduce una vida y devuelve nuevo estado"""
    nuevo_estado = estado_juego.copy()
    nuevo_estado["vidas"] -= 1
    
    if nuevo_estado["vidas"] <= 0:
        print(" Game Over")
    else:
        print(f" Perdiste una vida. Vidas restantes: {nuevo_estado['vidas']}")
    
    return nuevo_estado

def mostrar_estado_juego(estado):
    print(f"\nESTADO DEL JUEGO")
    print(f"Jugador:{estado['usuario']}")
    print(f"Puntuacion:{estado['puntuacion']:,}")
    print(f"Nivel:{estado['nivel']}")
    print(f"Vidas:{estado['vidas']}")
    print(f"Items:{','.join(estado['items']) if estado['items'] else 'Ninguno'}")

#Demostracion
print("SIMULACION DE JUEGO")

juego=crear_estado_juego_inicial("Ana")
mostrar_estado_juego(juego)
#eventos
juego=actualizar_puntuacion(juego,750)
juego=actualizar_puntuacion(juego,500)

mostrar_estado_juego(juego)
juego=perder_vida(juego)
juego["items"].append("Espada Magica")
juego=actualizar_puntuacion(juego,1200)
mostrar_estado_juego(juego)
```

```python
def crear_contador():
    cuenta=0

    def incrementar():
        nonlocal cuenta
        cuenta +=1
        return cuenta

    def obtener_cuenta():
        return cuenta

    def resetear():
        nonlocal cuenta
        cuenta=0

    return incrementar, obtener_cuenta,resetear

# Crear diferentes contadores independientes
print(" === CONTADORES INDEPENDIENTES ===")

contador1_inc, contador1_get, contador1_reset = crear_contador()
contador2_inc, contador2_get, contador2_reset = crear_contador()

# Usar primer contador
print(f"Contador 1: {contador1_inc()}")  # 1
print(f"Contador 1: {contador1_inc()}")  # 2
print(f"Contador 1: {contador1_inc()}")  # 3

# Usar segundo contador (independiente)
print(f"Contador 2: {contador2_inc()}")  # 1
print(f"Contador 2: {contador2_inc()}")  # 2

# Verificar que son independientes
print(f"Contador 1 actual: {contador1_get()}")  # 3
print(f"Contador 2 actual: {contador2_get()}")  # 2

# Resetear primer contador
contador1_reset()
print(f"Contador 1 despu茅s de reset: {contador1_get()}")  # 0
print(f"Contador 2 sin cambios: {contador2_get()}")      # 2
```
