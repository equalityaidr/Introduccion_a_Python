```python
edad = 16 

if edad >= 18: 
    print("Eres mayor de edad") 
    print("Puedes obtener licencia de conducir") 
else: 
    print("Eres menor de edad") 
    print(f"Te faltan {18 - edad} años para ser mayor de edad") 

```

```python
# Sistema de calificaciones 
nota = 85 
if nota >= 90: 
    calificacion = "A - Excelente" 
elif nota >= 80: 
    calificacion = "B - Muy bueno" e
elif nota >= 70: 
    calificacion = "C - Bueno" 
elif nota >= 60: 
    calificacion = "D - Regular" 
else: 
    calificacion = "F - Reprobado" 
    
print(f"Tu calificación es: {calificacion}")
```
