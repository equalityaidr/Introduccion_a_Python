```python
#Inventario de tienda
tienda={"Manzanas":{"precio":2.50,"stock":100,"categoria":"frutas"},
        "Leche":{"precio":3.20,"stock":50,"categoria":"lacteos"},
        "Pan":{"precio":1.80,"stock":30,"categoria":"panaderia"},
        "Yogurt":{"precio":4.50,"stock":25,"categoria":"lacteos"}
       }

print("ANALISIS DE INVENTARIO")

#keys-obtener todos los productos
productos=list(tienda.keys())
print(f"Productos disponibles:{productos}")
print(f"Total de productos:{len(productos)}")

#values-analisis de valores
precios=[info["precio"] for info in tienda.values()]
stocks=[info["stock"] for info in tienda.values()]

print(f"\nAnalisis de precios:")
print(f"Precio promedio:${sum(precios)/len(precios):.2f}")
print(f"Precio mas alto:${max(precios):.2f}")
print(f"Precio mas bajo:${min(precios):.2f}")

#items()-analisis producto por producto


print("REPORTE DETALLADO:")
valor_total_inventario=0

for producto,info in tienda.items():
    valor_producto=infor["precio"]*info["stock"]
    valor_total_inventario += valor_producto
    print(f"{producto.title()}:")
    print(f"Precio:${info["precio"]:.2f}")
    print(f"Valor en inventario:${valor_producto:.2f}")
```

```python
#Sistema de registro de actividades por dia

actividades_semanales {}

actividades_semanales.setdefault("lunes",[].append("Reunion de equipo")
actividades_semanales.setdefault("lunes",[].append("Programar Python")
actividades_semanales.setdefault("martes",[].append("Diseñar interfaz")
actividades_semanales.setdefault("miercoles",[].append("Testing")

print("AGENDA SEMANAL")

for dia, actividades in actividades_semanales.items():
        print(f"{dia.title()}:")
        for i, actividad in enumerate(actividades,1):
           print(f"{i}.{actividad}")
```

```python
#Sistema de tareas pendientes

tareas={
    "T001":{"descripcion":"Revisar codigo","prioridad":"alta","estimacion":2},
    "T002":{"descripcion":"Escribir documentacion","prioridad":"media","estimacion":4},
    "T003":{"descripcion":"Testing","prioridad":"alta","estimacion":3},
    "T004":{"descripcion":"Deploy","prioridad":"baja","estimacion":1}
}

print("GESTION DE TAREAS")

#pop
tarea_completada=tareas.pop("T001","Tarea no encontrada")
print(f"Tarea completada:{tarea_completada}")
print(f"Tareas resntantes:{len(tareas)}")

#popitem
if tareas: ultima_tarea_id, ultima_tarea_info=tareas.popitem()
    print(f"Ultima tarea removida:{ultima_tarea_id}-{ultima_tarea_info['descripcion']}")

print(f"\nTareas pendientes:")

for tarea_id, info in tareas.items():
    print(f"{tarea_id}:{info['descripcion']}(Prioridad:{info['prioridad']})")
```
