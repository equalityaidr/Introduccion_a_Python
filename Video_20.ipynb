```python
#Sistema de coordenadas de videojuego
jugador_posicion=(100,250)
enemigo_posiciones=[(50,100),(200,150),(300,200)]
print("posicion jugador:",jugador_posicion)
```

```python
#Calcular distancia (coordenadas no deben cambiar)
def calcular_distancia(punto1,punto2):
    x1,y1=punto1
    x2,y2=punto2
    distancia=((x2-x1)**2+(y2-y1)**2)**0.5
    return distancia
    
distancia=calcular_distancia(jugador_posicion, enemigo_posiciones[0])

print(f"Distancia al primer enemigo:{distancia:.2f}")

```
