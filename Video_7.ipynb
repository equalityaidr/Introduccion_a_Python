```python
# Las 4 operaciones básicas 
suma = 10 + 5 # 15 
resta = 10 - 3 # 7
multiplicacion = 4 * 7 
# 28 division = 15 / 3 # 5.0
print(suma)
```

```python
# Operaciones especiales
exponente = 2 ** 3   # 2 elevado a la 3 = 8
division_entera = 17 // 5  # División sin decimales = 3
residuo = 17 % 5     # Resto de la división = 2
print(exponente)
```

```python
#concatenacion
nombre = "Ana" 
apellido = "García" 
nombre_completo = nombre + " " + apellido 
print(nombre_completo) # "Ana García" 
# Más ejemplos 
saludo = "Hola " + nombre + ", ¿cómo estás?" 
print(saludo) 
# Concatenando con números (¡cuidado!)
edad = 25 
# Esto NO funciona: 
# mensaje = "Tengo " + edad + " años" # ERROR! 
# Esto SÍ funciona: 
mensaje = "Tengo " + str(edad) + " años" 
print(mensaje) 
```

```python
#f-strings
nombre = "Carlos" 
edad = 30 
salario = 2500.75 
# Forma antigua (complicada) 
mensaje1 = "Hola " + nombre + ", tienes " + str(edad) + " años" 
# Forma moderna (¡fácil!) 
mensaje2 = f"Hola {nombre}, tienes {edad} años y ganas ${salario}" 
print(mensaje2) 
```

```python
# ¡También funciona con operaciones! 
precio = 100 
descuento = 15 
mensaje3 = f"Precio original: ${precio}, con descuento: ${precio - descuento}" 
print(mensaje3)
```
