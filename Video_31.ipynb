```python
import matplotlib.pyplot as plt
import numpy as np 
 # Datos de ejemplo 
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun'] 
ventas = [120, 135, 148, 156, 142, 168] 
```

```python
#Grafico de linea
plt.figure(figsize=(10,6))
plt.plot(meses,ventas,marker='o',linewidth=2)
plt.title('Evolucion de Ventas 2024')
plt.xlabel('Mes')
plt.ylabel('Ventas (miles$)')
plt.grid(True,alpha=0.3)
plt.show()
```

```python
#grafico de barras
plt.figure(figsize=(8,6))
plt.bar(meses,ventas,color='skyblue',edgecolor='navy')
plt.title('Ventas por mes')
plt.ylabel('Ventas (miles$)')
plt.show()
```

```python
#histograma
datos_aleatorios=np.random.normal(100,15,1000)
plt.figure(figsize=(8,6))
plt.hist(datos_aleatorios,bins=30, color='red',alpha=0.7)
plt.title('Distribucion de datos')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

```python
sns.set_style("whitegrid") 
plt.figure(figsize=(12, 8)) 

# Datos de ejemplo ms ricos 
datos = pd.DataFrame({ 'mes': ['Ene']*30 + ['Feb']*30 + ['Mar']*30, 'ventas': np.concatenate([ np.random.normal(120, 20, 30), np.random.normal(135, 25, 30), np.random.normal(148, 15, 30) ]), 'categoria': np.random.choice(['A', 'B', 'C'], 90) }) 
```

```python
#grafico de caja(boxplot)
plt.subplot(2,2,3)
sns.boxplot(data=datos,x='mes',y='ventas')
plt.title('Distribucion de ventas por mes')
```

```python
#grafico de violin
plt.subplot(2,2,2)
sns.violinplot(data=datos,x='mes',y='ventas')
plt.title('Densidad de ventas por mes')
```

```python
#scatterplot con categorias
plt.subplot(2,2,3)
sns.scatterplot(data=datos, x='ventas',y=datos.index,hue='categoria',size='ventas')
plt.title('Ventas por Categoria')
```

```python
#Heatmap de correlacion
datos2=pd.get_dummies(datos,columns=['mes'],dtype=int)
numericos = datos2.select_dtypes(include=[np.number])
correlacion = numericos.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlacion , annot=True, cmap='coolwarm',fmt='.1f') 
plt.title('Correlacin Mes-Ventas') 
plt.show()
```

```python
# Los DataFrames tienen mtodos de plotting integrados 
ventas_df = pd.DataFrame({ 'fecha': pd.date_range('2024-01-01', periods=365), 'ventas': np.random.normal(1000, 200, 365) }) 

# Resample por mes y graficar 
ventas_mensuales = ventas_df.set_index('fecha').resample('M').sum() 
```

```python
#Multiples graficos en una linea
fig,axes=plt.subplots(2,2,figsize=(15,10))
ventas_mensuales.plot(ax=axes[0,0],kind='line',title='Linea')
ventas_mensuales.plot(ax=axes[0,1],kind='bar',title='Barras')
ventas_mensuales.plot(ax=axes[1,0],kind='area',title='Area')
ventas_df['ventas'].plot(ax=axes[1,1],kind='hist',bins=50,title='Histograma')
plt.tight_layout()
plt.show()
```

```python
#Dashboard simple
fig=plt.figure(figsize=(16,10))

#Datos simulados de una empresa
productos=['Laptops','Tablets','Smartphones','Auriculares']
ventas_producto=[450,320,680,150]
colores=['#FF6B6B','#4ECDC4','#45B7D1','#96CEB4']

# Grfico circular (pie chart) 
plt.subplot(2, 3, 1) 
plt.pie(ventas_producto, labels=productos, colors=colores, autopct='%1.1f%%') 
plt.title('Distribucin de Ventas por Producto') 

# Grfico de barras horizontal 
plt.subplot(2, 3, 2) 
plt.barh(productos, ventas_producto, color=colores) 
plt.title('Ventas por Producto (k)') 
plt.xlabel('Ventas') 

# Tendencia temporal 
fechas = pd.date_range('2024-01-01', periods=12, freq='M') 
tendencia = [400, 420, 450, 480, 460, 490, 510, 530, 520, 550, 580, 600] 
plt.subplot(2, 3, 3) 
plt.plot(fechas, tendencia, marker='o', color='#FF6B6B', linewidth=3) 
plt.title('Evolucin Mensual Total') 
plt.xticks(rotation=45) 

# Mtricas clave 
plt.subplot(2, 3, 4) 
plt.text(0.1, 0.8, f'Total Ventas\n${sum(ventas_producto):,}k', fontsize=16, fontweight='bold')
plt.text(0.1, 0.4, f'Promedio\n${np.mean(ventas_producto):.1f}k', fontsize=16, fontweight='bold') 
plt.axis('off') 
plt.title('KPIs Principales') 
plt.tight_layout() 
plt.show()
```
