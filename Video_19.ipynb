```python
#sistema de puntuaciones
puntuaciones=[]
print("SISTEMA DE PUNTUACIONES")
for ronda in range(1, 4):
    puntos = int(input(f"Puntuaci贸n ronda {ronda}: "))
    puntuaciones.append(puntos)
    print(f"Puntuaciones actuales: {puntuaciones}")

print(f"Puntuaciones finales: {puntuaciones}")
```

```python
#remove #pop #clear

animales=["perro","gato","chivo","pez","hamster"]

print("Lista original:",animales)

#remove
animales.remove("chivo")
print(animales)

#pop-devuelve el ultimo elemento
ultimo=animales.pop()
print(ultimo)
segundo=animales.pop(1)
print(segundo)

#clear
animales_copia=animales.copy()
animales_copia.clear()
print(animales_copia)
```

```python
#len count index
calificaciones=[85,90,85,78,92,85,88,85]

#len
print(f"Total de calificaciones:{len(calificaciones)}")

#count
print(f"Veces que aparece el 85:{calificaciones.count(85)}")

#index
posicion=calificaciones.index(92)
print(f"La calificacion 92 esta en la posicion:{posicion}")
```

```python
#min max sum
calificaciones=[85,90,85,78,92,85,88,85]

#min
print(f"Calificacion mas baja:{min(calificaciones)}")

#max
print(f"Calificacion mas alta:{max(calificaciones)}")

#sum
print(f"Suma total:{sum(calificaciones)}")
print(f"Promedio:{sum(calificaciones)/len(calificaciones):.2f}")
```

```python
#sort #sorted

numeros=[64,34,25,12,22,11,90]

print("Lista original:",numeros)

#sort
numeros.sort()

#orden ascendente
print("Orden ascendente:",numeros)
numeros.sort(reverse=True)

#sorted-crear una nueva lista ordenada
nombres=["Carlos","Ana","Bruno","Diana"]
print("nombres original:",nombres)

nombres_ordenados=sorted(nombres)
print(nombres_ordenados)
```
