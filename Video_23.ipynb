```python
#Creando conjuntos

numeros_unicos={1,2,3,4,5}
print("Conjunto de numeros:",numeros_unicos)

#Crear desde lista
lista_con_duplicados=[1,2,2,3,3,3,4,5,5]

conjunto_sin_duplicados=set(lista_con_duplicados)
print("Lista original:",lista_con_duplicados)
print("Conjunto sin duplicados:",conjunto_sin_duplicados)

#conjunto vacio
conjunto_vacio=set()
print("Conjunto vacio:",conjunto_vacio)

#conjunto desde string
palabra="programacion"
letras_unicas=set(palabra)
print(f"Letras unicas en '{palabra}':{letras_unicas}")
```

```python
#Gestion de estudiantes en diferentes cursos

python_estudiantes = {"Ana", "Bruno", "Carlos", "Diana", "Elena"}
javascript_estudiantes = {"Bruno", "Diana", "Franco", "Gaby", "Hugo"} 
sql_estudiantes = {"Ana", "Carlos", "Franco", "Ivan", "Julia"} 
 
print("ðŸŽ“ === ANÃLISIS DE INSCRIPCIONES ===") 
print("Python:", python_estudiantes) 
print("JavaScript:", javascript_estudiantes) 
print("SQL:", sql_estudiantes)

#Union
todos_estudiantes=python_estudiantes|javascript_estudiantes|sql_estudiantes
print(f"\nTodos los estudaintes:{todos_estudiantes}")
print(f"Total de estudiantes unicos:{len(todos_estudiantes)}")

#interseccion
python_y_js=python_estudiantes&javascript_estudiantes
print(f"\nEstudiantes Python y Javascript:{python_y_js}")

tres_cursos=python_estudiantes&javascript_estudiantes&sql_estudiantes
print(f"Estudiantes en los tres cursos:{tres_cursos}")

#diferencia
solo_python=python_estudiantes-javascript_estudiantes-sql_estudiantes
print(f"Solo en Python:{solo_python}")

#diferencia simetrica
python_xor_js=python_estudiantes^javascript_estudiantes
print(f"En Python o Javascript (no en ambos):{python_xor_js)}")
```

```python
#Sistema de tags para articulos
articulo1_tags = {"python", "programacion", "tutorial", "principiante"}
articulo2_tags = {"javascript", "web", "frontend", "tutorial"}
articulo3_tags = {"python", "data science", "pandas", "avanzado"}

print("ðŸ“° === SISTEMA DE TAGS ===")

#add()
nuevo_tag="machine learning"
articulo3_tags.add(nuevo_tag)
print(f"Tags articulos:{articulo3_tags}")

#update()
articulo1_tags.update(["basico","educacion"])
print(f"Tag actualizado articulo 1:{articulo2_tags}")

#remove()
try:
    articulo1_tags.remove("inexistente")
except KeyError:
    print("Tag no encontrado")

articulo1_tags.discard("inexistente")
articulo1_tags.discard("basico")
print(f"Despues de discard:{articulo1_tags}")

#Verificar membresia
tag_buscar="python"
print(f"\n'{tag_buscar}' esta en articulo 1?{'python' in articulo1_tags}")

#Encontrar tags populares
todos_los_tags=[articulo1_tags,articulo2_tags,articulo3_tags]
contador_tags={}

for tags in todos_los_tags:
    for tag in tags:
        contador_tags[tag]=contador_tags.get(tag,0)+1

tags_populares={tag for tag, count in contador_tags.items() if count>1}
print(f"Tags populares:{tags_populares}")
```
