```python
# Buscar un n√∫mero espec√≠fico
print("Buscando el primer numero divisible por 7 entre 1 y 100")

for numero in range(1,101):
    if numero %7==0:
        print(f"Encontrado!{numero} es divisible por 7") 
        break
        print(f"Revisando:{numero}")
print("Busqueda terminada")
```

```python
# Mostrar solo n√∫meros impares

print("Numero impares del 1 al 20")

for numero in range(1,21):
    if numero %2==0:
        continue
    print(f"Numero impar: {numero}")
```

```python
#Juego de adivinanzas

import random

print("JUEGO DE ADIVINANZAS")
print("Adivina el numero secreto entre 1 y 100")
print("Tienes 7 intentos. Buena Suerte!")

numero_secreo=random.randint(1,100)
max_intentos=7

for intento in range(1,max_intentos+1):
    try:
        guess = int(input(f"\nIntento {intento}/{max_intentos}: "))
        
        # Validar entrada
        if guess < 1 or guess > 100:
            print("‚ö†Ô∏è El n√∫mero debe estar entre 1 y 100")
            continue  # No contar este como intento v√°lido
        
        # Verificar adivinanza
        if guess == numero_secreto:
            print(f"üéâ ¬°FELICITACIONES! Adivinaste en {intento} intentos")
            print(f"El n√∫mero secreto era {numero_secreto}")
            break  # Salir del juego, gan√≥
        elif guess < numero_secreto:
            print("üìà El n√∫mero secreto es M√ÅS ALTO")
        else:
            print("üìâ El n√∫mero secreto es M√ÅS BAJO")
            
        # Dar pistas adicionales
        diferencia = abs(guess - numero_secreto)
        if diferencia <= 5:
            print("üî• ¬°Muy caliente!")
        elif diferencia <= 15:
            print("üå°Ô∏è Caliente")
        elif diferencia <= 30:
            print("‚ùÑÔ∏è Fr√≠o")
        else:
            print("üßä ¬°Muy fr√≠o!")
    except ValueError:
            print("Por favor ingresa solo numeros")
    continue
    else:
        print(f"\nSe acabaron los intentos!")
        print(f"El numero secreto era{numero_secreto}")

print("Gracias por jugar!")
```

```python
#Procesador de datos con filtros-procesador de calificaciones

print("PROCESADOR DE CALIFICACIONES")

#simular datos de estudiantes

estudiantes_datos=[("Ana",85),("Bruno",92),("Carlos",45),("Diana",78),("Elena",88),
                   ("Franco",35),("Gaby",95),("Hugo",67),("Iris",83),("Juan",91)]

print("Procesando calificaciones...")

print("Filtros: Solo estudiantes aprobados (>=60) y sin calidicaciones perfectas")

aprobados_procesados=0
total_suma=0

for nombre, calificacion in estudiantes_datos:
    if calificacion<60:
        print(f"{nombre} ({calificacion})-Reprobado, omitir")
        continue
    if calificacion==100:
        print(f"{nombre} ({calificacion})-Perfecto, omitir del promedio")
        continue
    if aprobados_procesados>=5:
        print(f"Limite de procesamiento alcanzado (5 estudiantes)")
        break

    aprobados_procesados +=1
    total_suma=+calificacion

print(f"{nombre}({calificacion})-Procesado #{aprobados_procesados}")

#calcular estadisticas

if aprobados_procesados>0:
    promedio=total_suma/aprobados_procesados

    print("RESUMEN")

    print(f"Estudiantes procesados: {aprobados_procesados}")
    print(f"Promedio del grupo: {promedio:.2f}")

else:
    print("No se procesaron estudiantes")
                   
```
