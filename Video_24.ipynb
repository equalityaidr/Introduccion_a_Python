```python
def calcular_area_rectangulo(largo,ancho):
    area=largo*ancho
    print(f"Rectangulo de {largo}x{ancho}")
    print(f"Area={area}")

#Usar funcion
calcular_area_rectangulo(5,3)
```

```python
def presentar_persona(nombre, edad, ciudad="No especificada",profesion="Estudiante"):
    print(f"PRESENTACION")
    print(f"Nombre:{nombre}")
    print(f"Edad:{edad}")
    print(f"Ciudad:{ciudad}")
    print(f"Profesion:{profesion}")

#Diferentes formas de llamar la funcion
presentar_persona("Ana",25)
presentar_persona("Bruno",22,"Madrid")
presentar_persona("Sophia",28,"Barcelona",profesion="Doctora")
```

```python
def evaluar_estudiante(nombre, calificaciones, mostrar_detalle=True):
    if not calificaciones:
        print(f"{nombre}:No hay calificaciones registradas")
        return
    
    # C√°lculos
    promedio = sum(calificaciones) / len(calificaciones)
    nota_maxima = max(calificaciones)
    nota_minima = min(calificaciones)
    
    # Determinar estado acad√©mico
    if promedio >= 90:
        estado = "üèÜ EXCELENTE"
    elif promedio >= 80:
        estado = "üåü MUY BUENO"
    elif promedio >= 70:
        estado = "üëç BUENO"
    elif promedio >= 60:
        estado = "‚ö†Ô∏è REGULAR"
    else:
        estado = "‚ùå NECESITA APOYO"
    
    # Mostrar resultados
    if mostrar_detalle:
        print(f"üìä === EVALUACI√ìN DE {nombre.upper()} ===")
        print(f"Calificaciones: {calificaciones}")
        print(f"Promedio: {promedio:.2f}")
        print(f"Nota m√°s alta: {nota_maxima}")
        print(f"Nota m√°s baja: {nota_minima}")
        print(f"Estado: {estado}")
        print()
    else:
        print(f"{nombre}: {promedio:.1f} - {estado}")

# Ejemplos de uso
estudiantes_datos=[("Ana Garcia",[85,90,88,92]),
                   ("Bruno Lopez",[78,82,75,80])]

print("EVALUACIONES DETALLADAS")

for nombre, notas in estudiantes_datos:
    evaluar_estudiante(nombre, notas)
    
print("RESUMEN RAPIDO")
for nombre, notas in estudiantes_datos:
    evaluar_estudiante(nombre, notas, mostrar_detalle=False)

```
